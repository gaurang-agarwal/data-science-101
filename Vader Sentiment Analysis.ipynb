```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./resources/moviereviews2.tsv', sep= '\t')
print(df.head())
```

```python
df.dropna(inplace=True)
```

```python
blanks = [];
for i,label,review in df.itertuples():
    if(isinstance(review,str) == False or label.isspace()):
        blanks.append(i)
df.drop(blanks,inplace=True);
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer();
```

```python

df['score'] = df['review'].apply(lambda review : sid.polarity_scores(review))
```

```python
df['compound_score'] = df['score'].apply(lambda s: s['compound'])
```

```python
print(df.head())
```

```python
df['feedback'] = df['compound_score'].apply(lambda cs : 'pos' if cs>0 else 'neg')
```

```python
from sklearn import metrics
# Print a confusion matrix
print(metrics.confusion_matrix(df['label'],df['feedback']))
print(metrics.classification_report(df['label'],df['feedback']))
print(metrics.accuracy_score(df['label'],df['feedback']))
```

```python
for row in df.iterrows():
    if(row[1]['label'] != row[1]['feedback']):
        print(f'{row[1]["review"]} <> {row[1]["score"]} <> {row[1]["compond_score"]}')
```
